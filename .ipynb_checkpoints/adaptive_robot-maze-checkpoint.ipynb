```python
def make_hex(n1,n2):
#     print(np.arange(n1))

    pointx = np.concatenate((np.arange(n1), np.arange(-0.5,n1+0.5,1)),axis = 0)
#     print(pointx)
    pointx = np.tile(pointx, n2)
    pointx = np.concatenate((pointx, np.arange(n1)),axis = 0)
#     print(pointx)

    pointy = np.zeros(n1)

    for i in range(1,n2*2+1):
        if i%2==1:
            pointy = np.concatenate((pointy,np.ones(n1+1)*np.sqrt(1-0.5**2)*i),axis = 0)
        else:
            pointy = np.concatenate((pointy,np.ones(n1)*np.sqrt(1-0.5**2)*i),axis = 0)

    i_points = np.array([pointx-np.mean(pointx), pointy[:len(pointx)]]).T
    return i_points
# rp = make_hex(2,1)
# rp[:,0] = (rp[:,0])*15
# fig = plt.figure(figsize = (8,8))
# ax = fig.add_subplot(111,aspect='equal') 

# ax.scatter(rp[:,0],rp[:,1])
# print(len(rp))

```

```python
def initialize():
    
    #initialize positions of everyone
    rs = np.zeros((N_s,2));
    raidi=N_s*D/2/np.pi;
    
    
    #rb[:,0] = 0.3*(np.random.rand(49)-0.5)+mod(range(49),7)-3.5;
    #rb[:,1] = 0.3*(np.random.rand(49)-0.5)+floor(np.arange(49)/7)+15;
    rb = make_hex(5,3)
#     rb = np.array([[-1,5]])
    rb[:,1] += max(rb[:,1])
    rb[:,0] += 3
    N_b = len(rb)
#     plt.scatter(rb[:,0], rb[:,1])
#     plt.show()
    rp = np.concatenate([rs, rb])# combined all movable particles
    #print(rp)
    
    rw = make_hex(3,4)
    rw[:,0] = rw[:,0].copy()*10
    rw[:,1] = rw[:,1].copy()*4
    rw[:,1] += -max(rw[:,1].copy())
#     rw = np.array([[-1,0],[7.5,0]])
    N_w = len(rw)

    thetas = (np.random.rand(N_s + N_b) * 2 - 1) * np.pi#np.zeros(N_s + N_b)+np.pi/2.# initial driving angle all point to right side
    vec_n = np.zeros(2)
    #print(vec_n[1])
    actf = np.zeros(N_s + N_b) #+c_active# active driving force initially 0
    
    return rs, raidi, rp, rw, thetas, vec_n, actf, N_b, N_w

def write_par():
    f = open(Datapath + "/parameters_run.txt", "w+")
    f.write('intrisic velocity of each active particle: \r\n' + str(c_g) + '\n')

    f.write('time step dt : \r\n' + str(dt) + '\n')
    f.write('Max iteration time : \r\n' + str(tau) + '\n')

    f.write('Number of the active particle :' + '\r\n' + str(N_b) + '\n')
    f.write('Number of the Wall  :' + '\r\n' + str(N_w) + '\n')

    f.write('Diameter of the active particle :' + '\r\n' + str(D) + '\n')
    f.write('Diameter of the wall : \r\n' + str(DW) + '\n')
    f.write('noise level : \r\n' + str(eta) + '\n')

    f.write('LJ strength : \r\n' + str(c_LJ) + '\n')
    f.write('MAX cutoff dis for the LJ between particle and wall : \r\n' + str(DW+FWR) + '\n')
    f.write('MIN cutoff dis for the LJ between particle and wall : \r\n' + str(0.8*DW) + '\n')

    f.write('Kspring : \r\n' + str(kspring) + '\n')
    f.write('cutoff dis from equilibrum dis for the spring between particle and particle : \r\n' + str(FR) + '\n')
    if c_active ==0:
        f.write('NOT active''\n')
    else:
        f.write('active''\n')
        if not_adap == False:    
            f.write('parameters discribing the adaptive force(A*(alp * ad_f / (1 + alp * ad_f))), A and alp:\r\n' + str(c_active) +',' + str(alp) + '\n')
    f.close()
```

```python
def make_movie(dat_hd_r, movname=[]):
    import subprocess

    imgname = dat_hd_r
    paths = dat_hd_r.split('/')
    datedir = ''
    for ii in range(len(paths) - 2):
        pn = paths[ii]
        datedir += pn + '/'
    if movname == []:
        movname = datedir + dat_hd_r.split('/')[-2] + '_xy0_1'

    print(imgname)
    print(movname)
    subprocess.call(['/Users/jiayiwu/simujupyter/lepm/ffmpeg', '-i',
                     imgname + '%*.png', '-filter:v', 'setpts=1*PTS',
                     '-vb', '20M', movname + '.mov', '-vcodec', 'libx264',
                     '-profile:v',
                     'main',
                     '-crf',
                     '1',
                     '-threads',
                     '0', '-r',
                     '100', '-pix_fmt', 'yuv420p'])
    
def plot_circles(fig,ax,x,y,r,c='r',cmap = matplotlib.cm.rainbow, ticks=[0,1]):
    patches = []
    for x1, y1 in zip(x, y):
        circle = plt.Circle((x1, y1), r)
        patches.append(circle)
    
    if isinstance(c, str):
        p = PatchCollection(patches,facecolor="None",edgecolor = c, alpha=1,linewidths=2)
        ax.add_collection(p)
        return ax
    
    else:   
        p = PatchCollection(patches, cmap=cmap, alpha=0.8)
        p.set_array(np.array(c))
     
    ax.add_collection(p)
    
    #cbar = fig.colorbar(p, ax=ax,ticks=ticks)
    p.set_clim(min(ticks),max(ticks))

def plot_active_force(c_active, alp, c_g):
    plt.plot(np.arange(0,10,0.1),c_active/c_g * (alp * np.arange(0,10,0.1) / (1 + alp * np.arange(0,10,0.1))))
    plt.ylabel('active velocity ($1/c_g$)', fontsize = 20)
    plt.xlabel('total velocity from last time v(t-1)', fontsize = 20)
#     plt.show()

    plt.savefig(Datapath + '/adaptive_force_profile.png')
    plt.tight_layout()
#     plt.close()

def visualization(not_adap):
    fig = plt.figure(figsize = (8,8))
    if not_adap:
        ax = fig.add_subplot(111,aspect='equal') 
    else:
        ax = fig.add_subplot(121,aspect='equal') 
    plt.title('time at %0.2f'%time+' with active = %0.2f '%(c_active) ,fontsize=20)

    plot_circles(fig,ax,rw[:,0],rw[:,1],DW-R,la.norm(FWfeel, axis = 1), cmap = matplotlib.cm.coolwarm, ticks=[0,100])
    if c_active ==0:
        plot_circles(fig,ax,rp[:,0],rp[:,1],R, c_active*actf)
    else:
        plot_circles(fig,ax,rp[:,0],rp[:,1],R, actf)
    plt.xlabel('force on wall = %d'%tot_Fwall,fontsize=20)
    ax.quiver(rp[:,0],rp[:,1], force[:,0], force[:,1], linewidth = 0.5)
    #print(force[:,0], force[:,1])

    plt.xlim(-10,10)
    plt.ylim(-40,15)
    if not_adap==False:
        a = plt.axes([0.6, 0.4, .2, .2])
        xx = np.arange(0,20,0.5)
        plt.plot(xx,(alp * xx / (1 + alp * xx)))

        plt.scatter(ad_f, actf, c=actf, cmap = matplotlib.cm.rainbow,
                    vmin=0, vmax=1, alpha = 1.)
        plt.colorbar()
        plt.ylabel('active velocity ')
        plt.xlabel('total velocity from last time v(t-1)')
        plt.tight_layout()
    return fig
        
```

```python
N_s = 0#skin active connected by spring

tot_Fwall = 0 # total force acting on wall

# N_b = 49#body particle
# N_w = 11#wall particles fixed at space;
dt = 0.005
tau = 40
eta = 0.1
sqdt = np.sqrt(dt) #noise 

R = 0.5#radius of beads
D = R * 2# diameter
DW = R + 4 * R

FR = R# cutoff range of force (repulsion if surface distance < R)
kspring = 2.

FWR = 2 * R
c_LJ = 1.
c_cutoff = c_LJ *(12*(DW/(0.8*DW)) ** (13))
print(c_cutoff)


alp = 1.5
#strengthA = 1.# strength of truncated active force
kspring = 2.

c_g = 2
las = 1.#[0,1] when las=1 the change of theta is instantaneous 
#kappa = 2 
#if =1, angle = the total force direction = actual motion
#if =2. angle = the vector sum of actual motion + active motion
```

```python
not_adap = False
c_active = 10.
c_g = 2.
kspring = 4.#c_g
las = 0.1#[0,1] when las=1 the change of theta is instantaneous 
if not_adap:
    print('its not adaptive')
# if c_active ==0:
#     name = '/Users/jiayiwu/experiment/soft_robot' + "/Not_active_c_g_vs_FWall_w_kspring_%02d.txt"%kspring

# else:
#     name = '/Users/jiayiwu/experiment/soft_robot' + "/active_%d_spring_%d_alp_%d_c_g_vs_FWall_w_change_active.txt"%(c_active,kspring,alp)
if las ==1:
    rootpath = '/Users/jiayiwu/experiment/soft_robot/maze/instantanous/'
else:
    rootpath = '/Users/jiayiwu/experiment/soft_robot/maze/continuous/'
if not os.path.exists(rootpath): os.mkdir(rootpath)
# f = open(name, "a")
# f.write('time' + ',')
# f.write('tot_Fwall' + ',')
# f.write('c_g' + ',')
# f.write('c_active' + ',')
# f.write('kspring' + ',')
# f.write('alp' + '\r\n')
# f.close()
```

```python

AVEFORCES = []
TIMES = []
for c_active in np.arange(5,20):
    if not_adap:
        print('its not adaptive')
    if las !=1:
        print('not instantanous')

#     for c_g in range(2,12,5):
    rs, raidi, rp, rw, thetas, vec_n, actf, N_b, N_w = initialize()

    #c_g = 2.# velocity downward
    Timenow = str(datetime.now().strftime("%Y_%m_%d_%H_%M_%S"))
    Datapath = rootpath +'c_act_%02d_c_g_%02d_k_%02d_'%(c_active,c_g,kspring)+str(Timenow)
    if not os.path.exists(Datapath): os.mkdir(Datapath)
    print(Datapath)
    if c_active !=0 and not_adap == False:
        plot_active_force(c_active, alp, c_g)
    write_par()
    tot_Fwall =0

    for time in np.arange(0,tau,dt):

        #initializefoce by the thermal noise force
        force = eta * (np.random.rand(N_s + N_b, 2) - 0.5) / sqdt#!!!!!!!!!!!!!!!!!!
        FWfeel = np.zeros((N_w,2))

        #loop around all movable particles:    
        #calculate repulsive force felt by any movalbe particle

        for p_id in range(0, N_s+N_b):
            #print(rp)
            x = rp[p_id, 0]
            y = rp[p_id, 1] 

            #%%%force with wall particles
            for i in range(0,N_w):
                vec_n[0] = -rw[i, 0] + x
                vec_n[1] = -rw[i, 1] + y
                dis = la.norm(vec_n)    # calculate distance
                if dis - DW < FWR:        # only count when within range

                    strength = np.min([c_cutoff,c_LJ *(12*(DW/dis) ** (13))])
                    #strength = c_rw * (1-dis+DW)^n_rep;

                    force[p_id, :] += strength * vec_n/dis
                    FWfeel[i,:] += strength * vec_n/dis

            tot_Fwall += sum(la.norm(FWfeel, axis = 1))

            #%%%force with other body+skin particle
            for i in range(N_s + N_b):    # all movable particles
                if i != p_id:        # but can't interact with self
                    vec_n[0] = -rp[i, 0] + x
                    vec_n[1] = -rp[i, 1] + y
                    dis = la.norm(vec_n)        # calculate distance
                    if dis - D < FR:
                        #print(dis, x, y, rp[i,0], rp[i,1])
                        strength = np.round(-kspring*(dis-D),2) #+ np.min([100,c_LJ * (12*(D/dis) ** (13) - 12 * (D/dis) ** (7))])                  
                        #strength = c_r * (1-dis+D)^n_rep;
                        force[p_id, :] = force[p_id, :] + strength * vec_n/dis

        force[:, 0] += c_active * (actf * np.cos(thetas))    
        force[:, 1] += c_active * (actf * np.sin(thetas)) - c_g

        phi = np.zeros(N_s + N_b)
        for i in range(N_s+N_b):
            aaa = force[i, :]
    #         if c_active !=0:

    #             aaa[0] += (kappa - 1) * actf[i] * np.cos(thetas[i])# this is vector sum of total force and active force
    #             aaa[1] += (kappa - 1) * actf[i] * np.sin(thetas[i])# this is vector sum of total force and active force

            angles = np.arctan2(aaa[1], aaa[0]) #+ eta*2*np.pi*np.random.rand(1)# vector aaa sets new active angle

            phi[i] = angles# here I can add modification of angle

        rp = rp + dt * force

        thetas = thetas*(1-las) + phi.copy()*(las)#  update my active angle to calculated from (5)
        if time ==0:
            oldforceamp = np.sqrt(force[:, 0] ** 2 + force[:, 1]**2)
            ad_f = oldforceamp
        else:
            ad_f = abs(np.max([(0.01*np.ones(len(oldforceamp)), oldforceamp)], axis=1))[0]
            actf = (alp * ad_f / (1 + alp * ad_f))
            if not_adap:
#                 print('its not adaptive')
                actf = np.ones(len(oldforceamp))

            oldforceamp = np.sqrt(force[:, 0] ** 2 + force[:, 1]**2)# calculate old force amplitude
            
        # datasaving and plotting
        if (time/dt)%10==0:

            savedir = Datapath+'/rp_force_data/'

            if not os.path.exists(savedir): os.mkdir(savedir)
            np.save(savedir+'/%04d'%(time/dt), np.array([rp, force]))
            fig = visualization(not_adap)
#             sys.exit()
            plotdir = Datapath+'/images/'
            if not os.path.exists(plotdir): os.mkdir(plotdir)
            plt.savefig(plotdir + '/%04d.png'%(time/dt))
#             plt.show()
            plt.close()            
            

        # check if it goes through the maze
        if max(rp[:,1])<-30:
            print('all particles go through at time', time)
            print('with the total force exerted on wall', tot_Fwall)
            print('c_g is', c_g)
#             if c_active ==0:
#                 f = open(name, "a")
#             else:
#                 f = open(name, "a")
#             f.write(str(time) + ',')
#             f.write(str(tot_Fwall) + ',')
#             f.write(str(c_g) + ',')
#             f.write(str(kspring) + ',')
#             f.write(str(c_active)+',')
#             f.write(str(alp) + '\r\n')
#             f.close()
            AVEFORCES.append(tot_Fwall/time)
            TIMES.append(time)
            make_movie(plotdir+'/', plotdir[:-8])
            
            break
```

```python
for time in np.arange(0,tau,dt):
    #(1)%%%%%%%%%%%%%%%%%%%%%%%%%%
    #initializefoce by the thermal noise force
    force = eta * (np.random.rand(N_s + N_b, 2) - 0.5) / sqdt#!!!!!!!!!!!!!!!!!!
    FWfeel = np.zeros((N_w,2))
    
    #(3)%%%%%%%%%%%%%%%%%%%%%%%%%%
    #loop around all movable particles:    
    #calculate repulsive force felt by any movalbe particle
    
    for p_id in range(0, N_s+N_b):
        #print(rp)
        x = rp[p_id, 0]
        y = rp[p_id, 1] 

        #%%%force with wall particles
        for i in range(0,N_w):
            vec_n[0] = -rw[i, 0] + x
            vec_n[1] = -rw[i, 1] + y
            dis = la.norm(vec_n)    # calculate distance
            if dis - DW < FWR:        # only count when within range
                
                strength = np.min([c_cutoff,c_LJ *(12*(DW/dis) ** (13))])
                #strength = c_rw * (1-dis+DW)^n_rep;

                force[p_id, :] += strength * vec_n/dis
                FWfeel[i,:] += strength * vec_n/dis
                
        tot_Fwall += sum(la.norm(FWfeel, axis = 1))

        #%%%force with other body+skin particle
        for i in range(N_s + N_b):    # all movable particles
            if i != p_id:        # but can't interact with self
                vec_n[0] = -rp[i, 0] + x
                vec_n[1] = -rp[i, 1] + y
                dis = la.norm(vec_n)        # calculate distance
                if dis - D < FR:
                    #print(dis, x, y, rp[i,0], rp[i,1])
                    strength = np.round(-kspring*(dis-D),2) #+ np.min([100,c_LJ * (12*(D/dis) ** (13) - 12 * (D/dis) ** (7))])                  
                    #strength = c_r * (1-dis+D)^n_rep;
                    force[p_id, :] = force[p_id, :] + strength * vec_n/dis
            
    force[:, 0] += c_active * (actf * np.cos(thetas))    
    force[:, 1] += c_active * (actf * np.sin(thetas)) - c_g
    
    phi = np.zeros(N_s + N_b)
    for i in range(N_s+N_b):
        aaa = force[i, :]
#         if c_active !=0:
            
#             aaa[0] += (kappa - 1) * actf[i] * np.cos(thetas[i])# this is vector sum of total force and active force
#             aaa[1] += (kappa - 1) * actf[i] * np.sin(thetas[i])# this is vector sum of total force and active force

        angles = np.arctan2(aaa[1], aaa[0]) #+ eta*2*np.pi*np.random.rand(1)# vector aaa sets new active angle

        phi[i] = angles# here I can add modification of angle
    
    rp = rp + dt * force
    #print(force)
    #sys.exit()

    thetas = phi.copy()#  update my active angle to calculated from (5)
    if time ==0:
        oldforceamp = np.sqrt(force[:, 0] ** 2 + force[:, 1]**2)
    else:
        ad_f = abs(np.max([(0.01*np.ones(len(oldforceamp)), oldforceamp)], axis=1))[0]
        actf = (alp * ad_f / (1 + alp * ad_f))
        oldforceamp = np.sqrt(force[:, 0] ** 2 + force[:, 1]**2)# calculate old force amplitude
    if max(rp[:,1])<-25:
        print('all particles go through at time', time)
        print('with the total force exerted on wall', tot_Fwall)
        print('c_g is', c_g)
        f = open('/Users/jiayiwu/experiment/soft_robot' + "/c_g_vs_FWall.txt", "a")
        f.write(str(time) + ',')
        f.write(str(tot_Fwall) + ',')
        f.write(str(c_g) + '\r\n')
        f.close()
        sys.exit()
        
    
    if (time/dt)%10==0:
        
#         print(time, tot_Fwall)
        savedir = Datapath+'/rp_force_data/'
        
        if not os.path.exists(savedir): os.mkdir(savedir)
        np.save(savedir+'/%04d'%(time/dt), np.array([rp, force]))

        
        fig = plt.figure(figsize = (8,8))
        ax = fig.add_subplot(111,aspect='equal') 
        plt.title('time at %0.2f'%time+' with active = %d (c_g)'%(c_active/c_g) ,fontsize=20)
        
        plot_circles(fig,ax,rw[:,0],rw[:,1],4*R,la.norm(FWfeel, axis = 1), cmap = matplotlib.cm.coolwarm, ticks=[0,100])
        plot_circles(fig,ax,rp[:,0],rp[:,1],R, phi[N_s:N_s + N_b])
        plt.xlabel('force on wall = %d'%tot_Fwall,fontsize=20)
        ax.quiver(rp[:,0],rp[:,1], force[:,0], force[:,1], linewidth = 0.5)
        #print(force[:,0], force[:,1])
        
        plt.xlim(-10,10)
        plt.ylim(-30,15)
        
        plotdir = Datapath+'/images/'
        if not os.path.exists(plotdir): os.mkdir(plotdir)
        plt.savefig(plotdir + '/%04d.png'%(time/dt))
        plt.close()

        #plt.show()
        #print(force)
        #print(phi)
        #sys.exit()
```
